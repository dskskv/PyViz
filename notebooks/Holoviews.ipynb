# Holoviews

*HoloViews is a Python library that makes analyzing and visualizing scientific or engineering data much simpler, more intuitive, and more easily reproducible*

Reference: http://holoviews.org/

### Interfacing with `pandas` and `seaborn`

Example taken from: http://holoviews.org/Tutorials/Pandas_Seaborn.html

```python
# Importing required libraries

import itertools
import numpy as np
import pandas as pd
import seaborn as sns
import holoviews as hv
```

```python
np.random.seed(22)
```

```python
hv.notebook_extension()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
%output holomap="widgets" fig="svg"
```

```python
%%opts Distribution (hist=False kde_kws=dict(shade=True))
d1 = 25 * np.random.randn(500) + 450
d2 = 45 * np.random.randn(500) + 540
d3 = 55 * np.random.randn(500) + 590

hv.Distribution(d1, label="Blue") *\
hv.Distribution(d2, label="Red") *\
hv.Distribution(d3, label="Yellow")
```

```python
%%opts Distribution (rug=True kde_kws={'color':'indianred', 'linestyle':'--'})
hv.Distribution(np.random.rand(10), vdims=['Activity'])
```
